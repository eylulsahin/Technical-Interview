## Question 4
Implement code to fill the missing data (impute) in daily_vaccinations column per country with the minimum daily vaccination number of relevant countries.  
Note: If a country does not have any valid vaccination number yet, fill it with “0” (zero)

```python
frames = []
for i in list(set(mydf['country'])):
    df_country = mydf[mydf['country']== i]
    if len(df_country) > 1:    
        df_country['daily_vaccinations'].fillna(df_country['daily_vaccinations'].min(),inplace = True)        
    else:
        df_country['daily_vaccinations'].fillna(mydf['daily_vaccinations'].min(),inplace = True)
    frames.append(df_country)    
    final_df = pd.concat(frames)

print(final_df.isnull().sum())
```

All Nan values are replaced. The new values are shown below.

```python
final_df = final_df.sort_index()
final_df.head(50)
```

Nan values are replaced with the country's minimum daily vaccination value for each country.

## Question 5
Implement code to list the top-3 countries with highest median daily vaccination numbers by considering missing values imputed version of dataset.

```python
country_mwedians = final_df.groupby(['country']).median()
country_mwedians['daily_vaccinations'].nlargest(3)
```

As it can be seen above, top-3 countries with highest median daily vaccinations are **United States, China and India**.
